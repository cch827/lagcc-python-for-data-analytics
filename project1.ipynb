```python
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
```

```python
df.replace('s', pd.NA, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.dropna()
```

```python
df.groupby(['School Name']).max().sort_values(by='Mean Score', ascending=False)
```

```python
df.rename(columns={'Regents Exam':'Subject','Total Tested':'Total Number of Tests'})
```

```python
import matplotlib.pyplot as plt

df_plot = df.groupby(['Year'])['Mean Score'].count().sort_values(ascending=False).reset_index()
df_plot.plot(kind='bar', x='Year', y='Mean Score')
plt.show()
```
